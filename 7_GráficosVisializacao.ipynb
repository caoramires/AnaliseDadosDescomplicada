# Preparando o ambiente

```python
# Preparando o ambiente para visualização
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Criando um dataset mais robusto para os exemplos
dados_completos = pd.DataFrame({
    'horas_estudo': [2, 3, 5, 1, 6, 7, 4, 8, 2, 5],
    'nota_final': [6.5, 7.0, 8.5, 5.0, 9.0, 9.5, 7.5, 10.0, 6.0, 8.0],
    'materia': ['Mat', 'Port', 'Mat', 'Port', 'Mat', 'Mat', 'Port', 'Mat', 'Port', 'Port']
})
```

## Histograma

```python
sns.histplot(dados_completos['nota_final'], kde=True)
plt.title('Distribuição das Notas Finais da Turma')
plt.xlabel('Nota Final')
plt.ylabel('Frequência (Nº de Alunos)')
plt.show()
```

## Gráfico de Dispersão

```python
sns.scatterplot(x='horas_estudo', y='nota_final', data=dados_completos)
plt.title('Relação entre Horas de Estudo e Nota Final')
plt.xlabel('Horas de Estudo por Semana')
plt.ylabel('Nota Final')
plt.show()
```

# Gráfico de Barras

```python
sns.barplot(x='materia', y='nota_final', data=dados_completos)
plt.title('Média de Nota por Matéria')
plt.xlabel('Matéria')
plt.ylabel('Média da Nota Final')
plt.show()
```

# Boxplot

```python
sns.boxplot(x='materia', y='nota_final', data=dados_completos)
plt.title('Dispersão das Notas por Matéria')
plt.show()
```
