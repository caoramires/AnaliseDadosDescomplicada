# Regressão Linear Simples

Ele tenta encontrar a melhor linha reta que descreve a relação entre uma única variável <br>
preditora (feature) e a variável alvo (target).

### Implementando um exemplo com Scikit-learn

```python
 # Importando as bibliotecas necessárias

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
import matplotlib.pyplot as plt
```

```python
 # Criando um DataFrame de Investimentos em Marketing

dados = pd.DataFrame({
    'investimento_marketing': [1000, 2000, 3000, 4000, 5000],
    'vendas': [50000, 65000, 80000, 92000, 110000]
})
```

```python
 # Criando um DataFrame de Investimentos em Marketing

dados = pd.DataFrame({
    'investimento_marketing': [1000, 2000, 3000, 4000, 5000],
    'vendas': [50000, 65000, 80000, 92000, 110000]
})

# 2. Separando features (X) e target (y)
# Scikit-learn espera que o X seja um array 2D, por isso os colchetes duplos.
X = dados[['investimento_marketing']]
y = dados['vendas']

# 3. Criando e treinando o modelo
modelo_regressao = LinearRegression()
modelo_regressao.fit(X, y)

# 4. Fazendo uma nova previsão

investimento_futuro = pd.DataFrame([[6000]], columns=['investimento_marketing']) # Cria um DataFrame com o nome de coluna correto
previsao_venda = modelo_regressao.predict(investimento_futuro)

print(f"Previsão de vendas para um investimento de R$ 6.000: R$ {previsao_venda[0]:.2f}")

```

```python
# Visualizando a linha de regressão
plt.scatter(X, y, color='blue', label='Dados Reais')
plt.plot(X, modelo_regressao.predict(X), color='red', linewidth=2, label='Linha de Regressão')
plt.title('Investimento em Marketing vs. Vendas')
plt.xlabel('Investimento (R$)')
plt.ylabel('Vendas (R$)')
plt.legend()
plt.show()
```

# Interpretando e avaliando o Modelo

```python
# Extraindo o coeficiente angular (o 'm' da nossa equação)
coeficiente_angular = modelo_regressao.coef_[0]

# Extraindo o intercepto (o 'b' da nossa equação)
intercepto = modelo_regressao.intercept_

print(f"--- A Fórmula do Nosso Modelo ---")
print(f"Coeficiente Angular (coef_): {coeficiente_angular:.2f}")
print(f"Intercepto (intercept_): {intercepto:.2f}")
print(f"\nEquação da Reta: Vendas = {coeficiente_angular:.2f} * Investimento + {intercepto:.2f}")
```

# Cálculo das métricas de avaliação

```python
# Pedindo ao modelo para prever os resultados para os dados de treino
vendas_previstas = modelo_regressao.predict(X)

# Calculando as métricas de erro
mae = mean_absolute_error(y, vendas_previstas) # 'y' são as vendas reais
r2 = r2_score(y, vendas_previstas)

print(f"\n--- Métricas de Avaliação do Modelo ---")
print(f"Erro Médio Absoluto (MAE): R$ {mae:.2f}")
print(f"Coeficiente de Determinação (R²): {r2:.4f}")
```
